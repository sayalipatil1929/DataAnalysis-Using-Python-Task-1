# a.DATA LOADING

```python
import pandas as pd
```

```python
df = pd.read_csv('student-mat.csv',sep=";")
```

```python
df.head(15)
```

# b.DATA EXPLORATION

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
print(f"Dataset dimensions: {df.shape}")
```

# c.DATA CLEANING

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.shape)
```

# d.DATA ANALYSIS QUESTIONS

# 1. What is the average score in math (G3)

```python
average_score = df['G3'].mean()
print(f"The average final grade (G3) is: {average_score:.2f}")
```

# 2.How many students scored above 15 in their final grade (G3)?

```python
students_above_15 = df[df['G3'] > 15].shape[0]
print(f"Number of students who scored above 15: {students_above_15}")
```

# 3.Is there a correlation between study time (study time) and the final grade (G3)?

```python
correlation = df['studytime'].corr(df['G3'])
print(f"Correlation between study time and final grade: {correlation :.2f}")
```

# 4.Which gender has a higher average final grade (G3)?

```python
average_grade_by_gender = df.groupby('sex')['G3'].mean()
print(average_grade_by_gender)
```

# e.DATA VISUALIZATION

# 1.histogram of final grades (G3).

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['G3'], kde=True, bins=15, color='red', edgecolor='black')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()
```

# 2.scatter plot between study time and final grade

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='studytime', y='G3', data=df, color='green')
plt.title('Scatter Plot of Study Time vs Final Grade (G3)')
plt.xlabel('StudyTime')
plt.ylabel('Final Grade (G3)')
plt.show()
```

# 3.bar chart comparing the average scores of male and female students.

```python
plt.figure(figsize=(8, 6))
average_grade_by_gender.plot(kind='bar' ,color=['blue','pink'])
plt.title('Average Final Grade (G3) by Gender')
plt.xlabel('sex')
plt.ylabel('Average Final Grade (G3)')
plt.xticks(rotation=0)
plt.show()
```
